# Python version

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

y = np.sin(x)

fig, ax = plt.subplots(fig_size=(8, 6))

ax.plot(x, y)

ax.set_title("График функции y = sin(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.set_xlim(-2*np.pi, 2*np.pi)
ax.set_ylim(-1.2, 1.2)

ax.grid()

plt.show()
```
