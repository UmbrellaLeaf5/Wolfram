# Построение графика функции, выраженной неявно

```
ContourPlot[Abs[x] ^ 2 + Abs[y] ^ 2 == 1, {x, -1.1, 1.1}, {y, -1.1,
   1.1}]
```

Параметризация

```
Manipulate[ContourPlot[Abs[x] ^ a + Abs[y] ^ a == 1, {x, -1.1, 1.1}, 
  {y, -1.1, 1.1}], {a, 0.1, 5}]
```

Распараллеливание

```
ContourPlot[Abs[x] ^ # + Abs[y] ^ # == 1, {x, -1.1, 1.1}, {y, -1.1, 1.1
  }]& /@ Parallelize[{0.2, 1, 2, 100}]
```
